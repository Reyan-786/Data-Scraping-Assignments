```python
import pandas as pd
list=[]
for i in range (1,100):
    list.append(driver.find_element(by ='xpath' ,value ='//*[@id="cl-competitors-table"]/div[2]/div/div/div/div/div['+str(i)+']').text)
```

```python
for i in list:
    i.split('\n')
    print(i)
```

```python
df= pd.DataFrame({'company name':[''],'.com level':[''],'traffic':['']})
for i in range(1,100):
    comp_name = driver.find_element(by ='xpath',value ='//*[@id="cl-competitors-table"]/div[2]/div/div/div/div/div['+str(i)+']/div[1]').text
    com_level =driver.find_element(by ='xpath',value  ='//*[@id="cl-competitors-table"]/div[2]/div/div/div/div/div['+str(i)+']/div[2]').text
    traffic =driver.find_element(by ='xpath',value ='//*[@id="cl-competitors-table"]/div[2]/div/div/div/div/div['+str(i)+']/div[5]').text
    length = len(df)
    df.loc[length] ={'company name':comp_name,'.com level':com_level,'traffic':traffic}
    
```
